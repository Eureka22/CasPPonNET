```python
edge = []
for i in range(len(nodes)):
    node = nodes[i]
    dist = []
    for j in range(len(nodes)):
        nodeto = nodes[j]
        if node[1] == nodeto[1]:
            dist.append(0)
        else:
            d = node[2]*nodeto[2]/((node[3] - nodeto[3])**2 + (node[4] - nodeto[4])**2) /1e10
            dist.append(d)
    edge.append(dist)
```

```python
deltat = 14
steps = 400/deltat
citynames = [item[1] for item in c]
p = []
for i in range(steps):
    d1 = deltat*i
    d2 = deltat*(i + 1)
    size = []
    for name in citynames:
        s = 0
        for record in o:
            if (record[4] > d1) and (record[4] <= d2) and (record[2] == 'Confirmed') and (record[1] == name):
                s = s + record[5]
        size.append(s)
    p.append(size)
p = np.array(p)
print p
print p.shape

plt.plot(p)

p = p.transpose()
```

```python
def update(lambda1,lambda2,beta,p,sigma,n,steps,deltat,mu):
    #print type(p)
    sum1 = 0
    sum2 = 0
    for i in range(n):
        for k in range(1,steps):
            t1 = 0
            for m in range(k):
                t1 = t1 +p[i,m]*math.exp(-beta*(k-m)*deltat)
            t2 = t1*p[i,k]
            t3 = 0
            for j in range(n):
                for m1 in range(k):
                    t3 = t3+p[j,m1]*sigma[i,j]*math.exp(-beta*(k-m1)*deltat)
            t4 = t1*deltat
            sum1 = sum1 + t2/(lambda1*t1 + lambda2*t3+1e-20) - t4 - mu*2*lambda1
            t5 = t3*p[i,k]
            t6 = t3*deltat
            sum2 = sum2 + t5/(lambda1*t1 + lambda2*t3+1e-20) - t6 - mu*2*lambda2
            #print t1
    return sum1,sum2
    
```
