```python
def getdata(filename):
    
    #filename = 'Liberia'
    
    fin = csv.reader(file('../data/'+filename+'.csv', 'rb'))
    data = []
    for line in fin:
        if line[0] == 'Country':
            head = line
        else:
            data.append(line)
    
    
    #print head
    #print data
    for item in head:
        print item, head.index(item)
    
    def fil(x):return x[3]!='' and x[6] == 'Patient database'
    
    data = filter(fil,data)
    
    def parsedate(x):
        date = x.split('to ')[1]
        start = '05 January 2014'
        return (parse(date) - parse(start)).days   
    
    
    def strtoint(s):
        if s == '':
            return 0
        else:
            return int(float(s))
    
    def getinfo(x):return [x[0],x[3],x[5],x[6],parsedate(x[7]),strtoint(x[12])]
    
    data = map(getinfo,data)
    
   
    
    cities = set([line[1] for line in data])
    
    def cityfilter(s):
        def f(x):
            return x[1]==s
        return f
    
    citydata = {}
    for city in cities:
        citydata[city] = filter(cityfilter(city),data)
    return cities,citydata
```

```python
def plotdata(data,countries,cases,statuses,cumus):
    def makefilter(country,case,status,cumu):
        def f(x):
            return (x[0] == country) and (x[1] == case) and (x[2] == status) and (x[3] == cumu)
        return f
    plt.figure(figsize=(12,8))
    for country in countries:
        for case in cases:
            for status in statuses:
                for cumu in cumus:
                    fdata = filter(makefilter(country,case,status,cumu),data)
                    day = [item[5] for item in fdata]
                    num = [item[4] for item in fdata]
                    #print day
                    #print num
                    
                    for i in range(len(day)):
                        for j in range(i):
                            if day[i] > day[j]:
                                day[i],day[j] = day[j],day[i];
                                num[i],num[j] = num[j],num[i];
                                
                    # sort
                    if day!=[] and num!=[]:
                        plt.plot(day,num, label=country + ' '+ case + ' ' + status + ' ' + cumu)
    plt.legend(loc=2)
    plt.savefig('_'.join(countries) + '_'+ '_'.join(cases) + '_'  +'_'.join(statuses) + '_'  +'_'.join(cumus) + '.png', dpi=300)
    plt.savefig('_'.join(countries) + '_'+ '_'.join(cases) + '_'  +'_'.join(statuses) + '_'  +'_'.join(cumus) + '.pdf',)
    
plotdata(data,['Liberia','Sierra Leone','Guinea'],['DEATHS'],['Total'],['Cumulative'])
plotdata(data,['Liberia','Sierra Leone','Guinea'],['CASES'],['Total'],['Cumulative'])
plotdata(data,['Liberia','Sierra Leone','Guinea'],['CASES','DEATHS'],['Total'],['Cumulative'])
plotdata(data,['Liberia','Sierra Leone','Guinea'],['CASES'],['Total'],['In past 21 days'])
```

```python
def filterdata(data,countries,cases,statuses,cumus):
    def makefilter(country,case,status,cumu):
        def f(x):
            return (x[0] == country) and (x[1] == case) and (x[2] == status) and (x[3] == cumu)
        return f
    for country in countries:
        for case in cases:
            for status in statuses:
                for cumu in cumus:
                    fdata = filter(makefilter(country,case,status,cumu),data)
                    day = [item[5] for item in fdata]
                    num = [item[4] for item in fdata]
                    #print day
                    #print num
                    
                    for i in range(len(day)):
                        for j in range(i):
                            if day[i] > day[j]:
                                day[i],day[j] = day[j],day[i];
                                num[i],num[j] = num[j],num[i];
    return day,num
```

```python
day1, num1 = filterdata(data,['Guinea'],['DEATHS'],['Total'],['Cumulative'])
day2, num2 = filterdata(data,['Liberia'],['DEATHS'],['Total'],['Cumulative'])
day3, num3 = filterdata(data,['Sierra Leone'],['DEATHS'],['Total'],['Cumulative'])

numdeath = []
if len(day1) == len(day2) and len(day1) == len(day3):
    for i in range(len(day1)):
        numdeath.append(num1[i] + num2[i] + num3[i])
```
