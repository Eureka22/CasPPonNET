```python
import numpy as np
import scipy
from sklearn import *
import matplotlib.pyplot as plt
import csv
from dateutil.parser import parse
```

```python
%matplotlib inline
```

```python
def getdata(filename):
    
    #filename = 'Liberia'
    
    fin = csv.reader(file('../data/'+filename+'.csv', 'rb'))
    data = []
    for line in fin:
        if line[0] == 'Country':
            head = line
        else:
            data.append(line)
    
    
    #print head
    #print data
    for item in head:
        print item, head.index(item)
    
    def fil(x):return x[3]!='' and x[6] == 'Patient database'
    
    data = filter(fil,data)
    
    def parsedate(x):
        date = x.split('to ')[1]
        start = '05 January 2014'
        return (parse(date) - parse(start)).days   
    
    
    def strtoint(s):
        if s == '':
            return 0
        else:
            return int(float(s))
    
    def getinfo(x):return [x[0],x[3],x[5],x[6],parsedate(x[7]),strtoint(x[12])]
    
    data = map(getinfo,data)
    
   
    
    cities = set([line[1] for line in data])
    
    def cityfilter(s):
        def f(x):
            return x[1]==s
        return f
    
    citydata = {}
    for city in cities:
        citydata[city] = filter(cityfilter(city),data)
    return cities,citydata
```

```python
cities,citydata = getdata('Sierra')
```

```python
for city in cities:
    b = [line[5] for line in citydata[city]]
    plt.figure()
    plt.plot(b)
```
