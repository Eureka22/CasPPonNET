```python
import numpy as np
import scipy
from sklearn import *
import matplotlib.pyplot as plt
import csv
from dateutil.parser import parse
import requests
from bs4 import BeautifulSoup
```

```python
def getdata(filename):
    
    #filename = 'Liberia'
    
    fin = csv.reader(file('../data/'+filename+'.csv', 'rb'))
    data = []
    for line in fin:
        if line[0] == 'Country':
            head = line
        else:
            data.append(line)
    
    
    #print head
    #print data
    #for item in head:
    #    print item, head.index(item)
    
    def fil(x):return x[3]!='' and x[6] == 'Patient database'
    
    data = filter(fil,data)
    
    def parsedate(x):
        date = x.split('to ')[1]
        start = '05 January 2014'
        return (parse(date) - parse(start)).days   
    
    
    def strtoint(s):
        if s == '':
            return 0
        else:
            return int(float(s))
    
    def getinfo(x):return [x[0],x[3],x[5],x[6],parsedate(x[7]),strtoint(x[12])]
    
    data = map(getinfo,data)
    
   
    
    cities = set([line[1] for line in data])
    
    def cityfilter(s):
        def f(x):
            return x[1]==s
        return f
    
    citydata = {}
    for city in cities:
        citydata[city] = filter(cityfilter(city),data)
    return cities,citydata
```

```python
def dict_union(train, wagon):
    for key, val in wagon.iteritems():
        if not isinstance(val, dict):
            train[key] = val
        else:
            subdict = train.setdefault(key, {})
            dict_union(subdict, val)


countries = ['Sierra', 'Liberia', 'Guinea']
abbr = {'Sierra':'SL', 'Liberia':'LR', 'Guinea':'GN'};
cityinfo = []
for country in countries:
    cities, citydata = getdata(country);
    for city in cities:
        cityinfo.append([abbr[country], city])

        
        
cities, citydata1 = getdata('Sierra')
cities, citydata2 = getdata('Liberia')
cities, citydata3 = getdata('Guinea')

citydata = citydata1
dict_union(citydata,citydata2)
dict_union(citydata,citydata3)
```

```python
def OnlyCharNum(s,oth=''):
    s2 = s.lower();
    fomart = 'abcdefghijklmnopqrstuvwxyz0123456789'
    for c in s2:
        if not c in fomart:
            s = s.replace(c,'');
    return s;

ccinfo = []
for info in cityinfo:
    url = 'http://www.geonames.org/search.html?q='+info[1]+'&country='+info[0]
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    lines = []
    if soup.body != None:
        if soup.body.div != None:
            if soup.body.div.table != None:
                lines = soup.body.div.table.find_all('tr')
    
    cinfo = []
    
    
    for line in lines:
        #print line.prettify()
        #print line.find_all('span').span
        if line == None:
            continue
        if line.td!=None:
            if line.td.small!=None:
                smalls = line.find_all('small')
                #print line.td
                if smalls[0].contents[0] == u'1':
                    #print info,smalls
                    cinfo.append(info[0])
                    cinfo.append(info[1])
                    if len(smalls) >= 3:
                        ss = smalls[len(smalls)-1].contents[0]
                        if ss.find('pop')!= -1:
                            pop = ss.split('ion ')[1]
                            #print pop
                            pop = OnlyCharNum(pop);
                            cinfo.append(int(pop))
                        else:
                            cinfo.append(0)
                    else:
                        cinfo.append('')
                    if line.span!=None:
                        for s in line.span:
                            cinfo.append(float(s.contents[0]))
                    else:
                        cinfo.append('')
                        cinfo.append('')
    print cinfo
    if cinfo!=[]:
        ccinfo.append(cinfo)
```
